### Question 2
- 자동제어를 통해 네이버 홈페이지에 접속하고 ‘크롤링’를 입력하여 검색하는        수집시스템을 구축하시오.

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
driver= wb.Chrome()
driver.get("http://naver.com")

search = driver.find_element_by_id("query")
search.send_keys("크롤링")
search.send_keys(Keys.ENTER)
```

### Question 3
- 자동제어를 통해 G마켓 홈페이지에 접속하고 1위부터 10위까지 조건에 맞는 수집프로그램을 만드시오.


```python
driver = wb.Chrome()
driver.get("http://corners.gmarket.co.kr/Bestsellers")
time.sleep(1)

productList= []
priceList=[]
categoryList =[]

for i in range(10) :
    img = driver.find_elements_by_css_selector("img.lazy")
    img[i].click()
    time.sleep(1)

    product = driver.find_element_by_css_selector("h1.itemtit").text
    productList.append(product)
    price = driver.find_element_by_css_selector("strong.price_real").text
    priceList.append(price)
    category = driver.find_element_by_css_selector("li.on>a").text
    categoryList.append(category)

    driver.back()
    time.sleep(1)


dic={"상품명":productList, "가격" : priceList, "카테고리":categoryList}
gmarket = pd.DataFrame(dic)
gmarket


```

### Question 4
- 멜론차트에서 1~100위까지 정보를 수집하는 프로그램을 작성하시오.


```python
h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36"}
res = req.get("https://www.melon.com/chart/", headers = h)
soup = bs(res.text,"lxml")

song = soup.select("div.ellipsis.rank01 > span > a")
singer = soup.select("div.ellipsis.rank02>span")

songList = []
singerList =[]
rankList =[]

for i in range(len(singer)) :
    songList.append(song[i].text)
    singerList.append(singer[i].text)
    rankList.append(i+1)
    
dic = {"노래제목" : songList, "가수" : singerList,"순위": rankList}
melon = pd.DataFrame(dic)

melon.set_index("순위", inplace= True)

melon.to_csv("멜론차트.csv", encoding ="euc-kr")
```

### Question 5
- 자동제어를 통해 한솥 메뉴 페이지에 접속하고, 전체 메뉴의 메뉴명, 가격 정보를 수집하시오.

```python
driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")
time.sleep(1)
try :
    for i in range(30) :
        btn = driver.find_element_by_css_selector("a.c_05")
        btn.click()
        time.sleep(1)
except : 
    print("더보기 종료")
    
title = driver.find_elements_by_css_selector("h4.h.fz_03")
price = driver.find_elements_by_css_selector("div > div.item-text > div > strong")

titleList=[]
priceList=[]
for i in range(len(title)):
    titleList.append(title[i].text)
    priceList.append(price[i].text)
    
dic={"상품명":titleList , "가격": priceList}
food = pd.DataFrame(dic)

food.to_csv("한솥메뉴.csv", encoding ="euc-kr")
```

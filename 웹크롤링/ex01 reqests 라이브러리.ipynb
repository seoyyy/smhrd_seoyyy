```python
#브라우저인척 속이기
#딕셔너리 형태로 바꿔주기 {} - 모든데이터에서 :콜론이 있다는건 키와 벨류값이 존재하다는것
h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36"}
```

```python
res = req.get("https://www.melon.com/", headers = h)
```

# 네이버페이지에서 블로그, 뉴스 글자 수집!

```python
# 1. 네이버 메인페이지 정보를 요청

res = req.get("http://naver.com")
res.text
```

```python
# 2. 뉴스, 블로그 등과 같은 글자만 수집
# 2-1. html 태그정보를 알려주기(어떤태그를 가지고오라고) - 컴퓨터는 한글 몰름

```

```python
# 3. 컴퓨터가 이해할 수 있는 html언어로 변형 ( 내가들고온 정보는 hmtl처럼 보이지만 그냥 텍스트형태임)
# beautifulSoup 라이브러리  - 얘는 어던 폴더에 들어가있어서 bs4공간을 통해서 임포트해주겟다고해야함
# 사실은 bs객체화로 만들어주는것 ( bs객체화를 진행)
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(res.text,"lxml")
```

```python
#저 숩데이터에서 어떤글자가담긴 태그를 가지고와주세요
txt = soup.select("a.nav")  
#-> a태그들 중에 클래스네임이 nav인걸가져와라
# 필요한 요소를 검색해서 수집하는 명령 .select("선택자") ★ -시험
```

```python
# 반환받은 리스트 데이터중에 첫번째 데이터만 가지고올때 select_one("선택자") ★
soup.select_one("a.nav")
```

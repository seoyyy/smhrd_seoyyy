# 인덱싱과 슬라이싱


```python
# 문자열 인덱싱

s4='My name is SY'
print(s4[3])
print(s4[1])
print(s4[-1])
```

```python
# 문자열 덧셈 가능
print(s4[3]+s4[4]+s4[5]+s4[6])
```

```python
# 문자열 슬라이싱

print(s4[0:4]) #4전까지만 나옴. 인덱스 번호 0,1,2,3 까지나온다.
# 앞에있는 인덱스번호는 포함, 뒤에있는 인덱스번호의 -1까지 나옴.

#0부터 잘라내고 싶을때 
print(s4[:4])
print(s4[3:])
print(s4[-5:]) #음수인덱스 활용해서 마지막까지 가져오고싶을때
```

```python
# %기호 포매팅 -> %d를 활용해서 정수 대입
month = 3
day = 23
s6 = "오늘은 %d월 %d일 80%%입니다."%(month,day)
print(s6)
```

### 문자열 관련 함수



```python
s7 = "My name is Seoyoung"

# 문자열에 포함된 문자 개수 세기
print(s7.count('e'))
print(s7.count('Seoyoung'))

# 문자 위치 알려주기(가장먼저발견된s의 위치를 알려줌)
print(s7.index("s"))
print(s7.find("s"))
# index는 내가찾는값이 존재하지 않으면 오류, find는 -1 출력

# 대소문자 변환
print(s7.upper())
print(s7.lower())

# 문자열 바꾸기
print(s7.replace("Seoyoung","HanSeoyoung"))
print(s7.replace)

# 문자열 자르기
print(s7.split(","))
```

# Matplotlib 라이브러리

- 데이터를 가지고 그래프를그리고 시각화를 할 수 있는
- 한눈에 알아볼 수 있도록 도표나 차트 등으로 정리하는 것
- -> 시각화함으로써 데이터의 특징이나 분석을 할 수 있고 인사이트를 얻을 수 있음

#### 그래프를 표시해주는 라이브러리
- scatter - 산점도
- bar - 막대그래프 (범주데이터를 나타낼때)
- pie - 원형그래프
- histogram - 도수분포표( 연속적인 수치데이터를 나태낼때)

```python
#plot 함수 - 일직선으로 된 line plot 그려주는 함수
#점을 찍고 점을 이어서 그래프를 선으로 그려주는 (쌍이맞지않으면 그래프를 그릴 수 없음)
y = [2,4,6]

plt.plot(y) #그래프 그리기
plt.show() #그래프 출력
```

```python
x =[1,2,3]

y=[2,4,6]

plt.plot(x,y)
plt.show()
```

```python
# 스타일 바꿔줄 수 있음
# linestyle='' - 그려지는 line의 스타일을 지정하는 속성 -> ls
x =[1,2,3]

y=[2,4,6]

plt.plot(x,y , linestyle='--')
plt.show()
```

```python
x = [2,4,6,8]
y = [10,13,15,20]
y2 =[10,20,17,15]

plt.plot(x,y)
plt.plot(x,y2 , ls =':')
plt.show()
```

```python
# marker 지정
plt.plot(x,y, marker='o')
plt.show()
```

```python
#라인의 두께지정 - linewidth 
# lw
plt.plot(x,y, linewidth =5)
plt.show()
```

```python
#색상지정가능 color
plt.plot(x,y, color ='r')
plt.show()
```

```python
#rgb 컬로도 사용가능 !
plt.plot(x,y, color ='#ea0e85', marker='o')
plt.show()
```

```python
# marker face color or mfc
#마커의 색상 따로 지정
plt.plot(x,y, color ='#ea0e85', marker='o', markerfacecolor='yellow')
plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y,linestyle = '--',lw =5,ms= 15, mew=5, marker='o', mec ='green',markerfacecolor='red')
```

```python
# 범위 지정( 그림의 영역 지정) xlim(,) ,ylim(,)
plt.plot(x,y,linestyle = '--',lw =5,ms= 15, mew=5, marker='o', mec ='green',markerfacecolor='red')
plt.xlim(-2,8)
plt.ylim(-4,12)
plt.show()
```

```python
# 축에있는 숫자를 tick 이라고함
#tick 설정 xtick([]) , ytick([])
plt.plot(x,y,linestyle = '--',lw =5,ms= 15, mew=5, marker='o', mec ='green',markerfacecolor='red')
plt.xlim(-2,8)
plt.ylim(-4,12)
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

```python
# grid() 설정 - 모눈종이
plt.plot(x,y,linestyle = '--',lw =5,ms= 15, mew=5, marker='o', mec ='green',markerfacecolor='red')
plt.grid()
plt.show()
```

```python
# 여러 그래프 동시에 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, color ='r')
plt.plot(y, color ='k')
plt.plot(z, color= 'm')
plt.show()
```

```python
plt.plot(x, color ='r')
plt.show()

plt.plot(y, color ='k')
plt.show()

plt.plot(z, color= 'm')
plt.show()

#show를 통해 그래프를 보여주면 새도화지에 다시그려야함.
```

```python
#범례 표시

#그래프에 이름 달아주기 label
plt.plot(x,label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend()  # ->이름표주기 (범례)
plt.show()
```

```python
#한글 넣어주려면 한글을지원할수있는 폰트로 지정해야함 
#쥬피터는 한글지원이안됌, 이폰트가 한글을 지원하지않음.
plt.plot(x,label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')
plt.legend()
plt.show()
```

```python
# 폰트설정
from matplotlib import rc
rc('font',family='Malgun Gothic')

plt.plot(x,label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')

plt.legend()
plt.show()
```

```python
# xlabel : x축 이름표시
# ylabel : y축 이름표시
# title : 그래프 이름
plt.plot(x,label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')

plt.xlabel('x축')
plt.ylabel('y축',rotation=0)  #글자가 누워있으니 세워주기 rotation=90이 기본값
plt.title('x,y,z축 그래프')

plt.legend()
plt.show()
```

##### 차대차 사건 중 죽거나 다친 사람이 많은 발생지시도 알아보기

```python
#27개의 컬럼 보이게 하기 
pd.set_option('display.max_columns', 27)
```

```python
# 1. 차대차 사건중 (전체데이터중 차대차 사건만 가지고오기)
# 2. (1) 죽거나 다친 사람이 많은 (2)발생지 시도
# 3. 시각화
data
```

```python
# 1. 차대차 사건중
#불리언 인덱싱 ('차대차'인지 판별)
#사고유형_대분류 컬럼을가져와서 거기서 '차대차' 인데이터를 가진것만 가져오기
data['사고유형_대분류'] == '차대차'
```

```python
#전체데이터에서 '차대차'분류를 가진 데이터들만 인덱싱해서 가져오기
data[data['사고유형_대분류'] == '차대차']
```

```python
acc = data[data['사고유형_대분류'] == '차대차']
```

```python
# 그 중에서 2. (1) 죽거나 다친 사람이 많은->'사상자수' (2)'발생지시도' 컬럼 가지고오기
```

```python
acc[['사상자수','발생지시도']]
# acc.loc[:,['사상자수':'발생지시도']] 쓸때는 행열을 어떻게 나누지? ,로나눔 acc.loc(행,열)
```

```python
#각 시도에 사고가 얼마만큼 발생했는지 세주기
acc['발생지시도'].value_counts()
#but 시도별 사상자수 (심함정도) 파악힘듬
```

```python
#그걸 파악하기위해
# .groupby()
#같은 값끼리(그룹화) 묶어 통계 연산 값을 보여줌
#발생지시도별로 그룹화시킬거야 그리고 더해줄거야
acc.loc[:,['사상자수','발생지시도']].groupby('발생지시도').sum()
```

```python
acc.loc[:,['사상자수','발생지시도']].groupby('발생지시도')
#groupby는 연산과 같이쓰이지않으면 그냥 내부적으로 그룹화만시키고 외부적으로그걸 확인할수없음. 단지정보일뿐
#연산함수를 만나야 최종으로 통계 연산을 그룹별로 시킴!
#이차원데이터프레임으로 나타남
```

```python
acc_result = acc.loc[:,['사상자수','발생지시도']].groupby('발생지시도').sum()
```

```python
plt.plot(acc_result)
```

```python
x = acc_result.index
```

```python
y = acc_result['사상자수'].values
```

```python
#3.도식화!! 그래프그리기

#그래프 크기조절
plt.figure( figsize =(9,5))


plt.bar(x,y)

plt.xlabel('지역별')
plt.ylabel('사상자수',rotation=0)  #글자가 누워있으니 세워주기 rotation=90이 기본값
plt.title('지역별 사망교통사고 건수')


plt.show()
```

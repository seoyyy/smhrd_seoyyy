# 딕셔너리

- key값은 변할 수 없는 값
- value값은 변할 수 있는 값
- 딕셔너리는 key:value 한쌍으로 맵핑되어져 있는 자료구조
- 만들고 한번 불러줘야 사용 가능 ?? 변수명만들때 공백도 인식하는지 ?? 


딕셔너리 선언

```python
a = {'name':'김희연'}  #요소가 1개 담긴 상황
#네임이라는 키값에 김희연이라는 벨류값을 맵핑한다(집어 넣는다)

print(a)


b={'a':['양서린','최보현','하종현'],
  'b':['김주란','김순기','박호진']}   # 요소가 2개 담긴 딕셔너리

print(b)

c = {1:36 ,2:45, 3:100}
print(c)
```

```python
# 딕셔너리 자료형 확인
print(type(a))  # dict
print(type(b))  # dict
```

딕셔너리 관련 함수
- keys() : 키값들 확인 함수
- values() : 벨류값들 확인하는 함수
- items() : 키와 벨류 함께 확인하는 함수


```python
b.keys()
```

```python
b.values()
```

```python
b.items()
```

for-dict (for과 함께 활용해보기) 

```python
# 인덱스번호사용한 for문
for i in b:
    print(i)
```

```python
for i in b.values():
    print(i)
```

```python
# 딕셔너리 데이터 접근 딕셔너리[키값]
for i in b:
    print(b[i]) #딕셔너리 집근 방법 i == 키값들 ..!
```

```python
for i in b:     # 딕셔너리이기때문에 키값을 통해 접근했다가 선언한 딕셔너리안에 리스트형식이 있으니 그 리스트형식은 인덱스 번호로 접근가능
    print(b[i][1])
    
    # b딕셔너리 -> b['키']
    # b['키']->리스트[인덱스번호]
```

```python
# 박호진 정보 출력해보기
# b 사용
# b.values()
# b.keys()

for i in b.keys():
    print(b[i][2])
```

```python
for i in b.values():
    print(i[2])
```

```python
dict2 = [{'name':'주란','birth':'0212'},
        {'name':'지영','birth':'0211'},
        {'name':'호진','birth':'0322'}]   #이런식으로 리스트안에 딕셔너리를 넣어버릴수도잇음
#리스트 안에 딕셔너리를 하나의 요소처럼 사용할 수 있다.

len(dict2)
```

```python
# for문 활용하여 생일 정보들만 출력해보기 !(리스트를 for문으로 돌리는거임)
for i in dict2:
    print(i)  # 지금 i 는 딕셔너리임
    
    # i 딕셔너리 생일 접근 어덯게?

for i in dict2:
    print(i['birth'])
    # i 자체가 딕셔너리 이기 때문에 
```

# 최종 마무리

- {key:value}   ==>딕셔너리
- 데이터를 저장할 때 키 값을 통해서 데이터의 정보를 명확하게 입력한다. 
- 키값은 변할 수 없음 , value는 변할 수 있음. ★★★
- value는 수정, 삭제, 추가 가능
- 수정방법 : 딕셔너리['기존키의이름'] = '새로운 데이터'
- 추가방법 : 딕셔너리['새로운키'] = '데이터'
- 삭제방법 : del 딕셔너리['키']
- 데이터에 접근 : 딕셔너리['키']

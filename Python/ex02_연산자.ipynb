```python
num5 = input("정수를 입력하세요>>")
num6 = input("정수를 입력하세요>>")
num7 = int(input("정수를 입력하세요 >>")) #이렇게 바로 input에 int 씌어줄수도있음
print(int(num5)+int(num6))
print(int(num5)-int(num6))
print(int(num5)*int(num6))
print(int(num5)/int(num6))

res1 = int(num5)+int(num6)
res2 = int(num5)-int(num6)
res3 = int(num5)*int(num6)
res4 = int(num5)/int(num6)
print(f"더하기 결과 : {int(num5)+int(num6)}")
print(f"빼기 결과 : {res2}")
print(f"곱하기 결과 : {res3}")
print(f"나누기 결과 : {res4}")

print("곱하기 결과 : {}".format(int(num5)*int(num6)))
```

# 삼항연산자

#  a if 조건식 else b


```python
score = 80
"합격" if score >= 60 else "불합격"
```

```python
print("합격" if score>= 60 else "불합격")
```

```python
a = int(input("정수입력>>"))
b = int(input("정수입력>>"))

res = f'a:{a}' if a>b else f'b:{b}'

print(res)
```

```python
c = int(input("정수 입력 >>"))

res1 = "짝수" if c%2==0 else "홀수"
print(f"{c}는(은) {res1}입니다.")
```

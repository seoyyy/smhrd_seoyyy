```python
# 함수에 대한 설명 달아주기
# docsting 독스트링
#""" 설명 """

def cal(num1,num2,op):
    """ 덧셈과 뺄셈을 계산하는 함수 !!"""
    if op == '+':
        result = num1 + num2
    elif op == '-': 
        result = num1 - num2
    return result


#  if op == '+':
#        return num1 + num2
#    elif op == '-': 
#       return num1 - num2




num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))
op = input("연산자 입력(+,-) >>")

result = cal(num1, num2, op)
print("결과 : {}".format(result))
```

- 키워드 : * (아스타리카)


```python
def 함수명(*매개변수)
    실행문장
    return 반환값
```

```python
def add(*num):
    print(num)  # 반환 X 단순출력임
    
    
```

return 개념정리
- return 언제나 반환값이 하나이다.
- 함수 안에서 return을 만나게 되면 함수는 끝난다.

```python
def add_sub(num1,num2):
    return num1+num2, num1-num2
```

```python
result1,result2 = add_sub(10,7)
print(result1,result2)
```

```python
add_sub(10,7)  # 우린 이걸 튜플이라고 하기로했슴. 결과가 하나(튜플하나)라고 볼수있음.

a,b = (10,7)
print(a,b)
```

함수의 기본 모형 4가지

```python
# 1. 매개변수 ㅇ, 리턴 ㅇ
def hi(x):
    return f'{x}아 안녕~'

hi = hi('지영')
print(hi)
```

```python
# 2.매개변수 ㅇ, 리턴X    ---> 리턴이 없어서 데이터가 넘어오지않고 단지 출력만시켜놔서 변수에 들어가지 않음
def hi2(name):
    print(f"{name}아 안녕~")
    
hi = hi2('지영')
print(hi)
```

```python
# 3.매개변수x ,리턴ㅇ
def hi3():
    return "지영아 안녕!!"


result = hi3()
print(result)
```

```python
# 4. 매개변수X, 리턴X
def hi4():
    print('안녕하세요~')
    
hi4()
```

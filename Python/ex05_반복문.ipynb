# 반복문



```python
num1 = num2 = -1

while num1!=0  or num2 !=0:
    num1 = int(input("첫 번째 정수 입력 >>"))
    num2 = int(input("두 번째 정수 입력 >>"))
    res = num1+num2
    if num1!=0  or num2 !=0:
        print(f"두 정수의 합 : {res}")
print("프로그램 종료")
    
    

```

```python
# in 뒤에는 문자열 , 리스트, 튜플, range함수 가 올수 있음

for i in range(1,16,1):
    print(i, end = ' ')
    
#값이 가로로 나오게 하는 방법
#  함수에 대한 정보 보기위해 shift + tab 누르면 end 가 나옴. end='\n' (이건 개행-줄바꿈) 이게 기본값이라 세로로나옴
# end = ' ' 로 바꿔주면 됨 (sort함수의 reverse 처럼)
```

```python
# 1씩 증가시키고싶을때 증가값1은 생략가능/ 0부터 시작할때도 0은 생략가능(단 매개변수가 두개만 있을시에는 반드시 시작값, 종료값을 넣어줘야한다. 
# 즉 증가값이 1일경우만 생략가능)
# range()안에 1부터 100까지 나오고싶을때 +1해서 써줘야함.
for i in range(1,101):
    print(i, end = ' ')
```

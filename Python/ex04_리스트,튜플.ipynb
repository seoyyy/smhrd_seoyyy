# 리스트


```python
#리스트 생성
#어떤게 들어있는지 모를때는 비어있는 대괄호로 놔둬도 가능.
li1 = []
print(len(li1))

li2 =[1,2,3]
print(len(li2))

#문자열담기
li3=["a","b","c"]
print(len(li3))

#숫자 문자열 한꺼번에 섞어 넣을 수도 있음
li4 =[1.2,3,'a','b']
print(len(li4))

#list 안에 list 넣기
li5 = [1,2,['a','b',3]]
print(len(li5))
```

```python
#리스트안에있는 내가 원하는것만 꺼내서 쓰기 =>인덱싱,슬라이싱
#리스트 안에 보이진않지만 순서대로 숫자가 부과되어있음. 그 인덱스번호 이용

list1= [2,5,7,9,10]
print(list1[0])
print(list1[2]+list1[-1])
```

```python
# list 인덱싱
list2 =[1,2,3,['a','b','c']]

temp = list2[3]
print(temp)
#리스트2 안에 있는 리스트의 b를 가져오고싶을때 
print(temp[1])
print(list2[3][1])
#list2에 접근해서 list2의 3번 인덱스에서 1번 인덱스 = b
```

```python
# list 슬라이싱-> 리스트명[start인덱스:end인덱스]
#마지막까지 가져오고싶을때 end인덱스생략 가능 && end 인덱스 번호는 포함 안됌 -1된숫자의 인덱스가옴

list3=[1,8,7,3,5]
print(list3[1:3])
print(list3[:2])
print(list3[3:])

#슬라이싱으로 하면 리스트타입으로 나오고 인덱싱으로하면 정수형태로 나온다는 차이가 있음.
print(list3[3:4])
print(list3[3])

print(type(list3[3]))
```

```python
#리스트 더하기
list4 =[1,2,3]
list5=[3,4,5,6]
print(list4+list5)
print(list5+list4)
```

```python
# 리스트 값 수정하기
list6 = [1,3,4,7,2,8]
list6[1]= 5
print(list6)

list6[1:4]=[6]
print(list6)
```

```python
list7 =[1,2,3,4,5]
#앞에서 인덱싱으로 접근햇냐 슬라이싱으로 접근햇냐에 따라서 리스트의데이터가 추가변경에있어서 낱개로들어가냐 리스트로 들어가냐가 달라짐

#리스트 array 숫자 5를 7로 바꾸기
list7[-1]=7
print(list7)

#숫자 2를 리스트['a',b','c']로 바꾸기 (인덱싱으로 접근해서)
list7[1]=['a','b','c']
print(list7)

# 숫자 4,7을 'd','e','f','g'로 바꾸기 (슬라이싱으로 접근해서 가능)

list7[3:]=['d','e','f','g']
print(list7)

#숫자 1을 'h','i','j'로 바꾸기
list7[0:1]=['h','i','j']
print(list7)
```

```python
#리스트 깂 추가하기(.append = 맨 마지막 위치에 데이터가 추가된다)
list8 = [0,1,2,3,4]
list8.append(5)
print(list8)

list8.append(6)
```

```python
# 리스트 값 삽입(insert(인덱스번호,넣을 값))
list9 = [0,1,2,3,4]
list9.insert(1,9)

print(list9)

list9.insert(1,[9,3])
print(list9)
```

```python
# 값 삭제  ( delete와 remove둘다  있음)  ---> 둘의 차이는 ? 
list10 = [0,1,2,3,4,5]
del list10[3]
print(list10)


list10.remove(1)
print(list10)
```

```python
#list 순서뒤집기
list11=[9,77,13,51]
list11.reverse()
print(list11)
```

```python
#리스트 정렬
list12 =[9,77,13,51,100,3]
list12.sort(reverse=True)
#reverse라는 매개변수의 기본값은 false라서 true로바꿔주면ㄷ 뒤집어짐

#list12.reverse()
list12
```

```python
#리스트 값 위치 반환
list13 = ['a','b','c','d','e']
list13.index('c')
```

# 튜플



```python
#튜플 생성  
# ->튜플은 외부에서 넘어와서 그안에 값이 변경이 되지않고 보존되었으면 할때 사용 (얘도 인덱싱,슬라이싱 가능/ but insert, delete등 수정은 안댐)

t1 = (1,2,3,4,5)
t2=('a','b','c')
t3=(1,2,3,'a','b')
t4=(1,2,3,('a','b'))
#튜플에 한개의 값만 넣고 싶을때는 ,콤마 넣어주기
t6=(1,)

print(t1)
print(len(t1))
print(t2)
print(len(t2))
print(t3)
print(len(t3))
print(t4)
print(len(t4))
print(t6)
```

```python
s = "Hi, My name is Mihee"
ans = input("검색할 문자를 입력하세요  :")

if ans not in s :
    print(f"{ans}는 문자열에 들어있지 않습니다.")
else:
    print(f"'{ans}'는 {s.count(ans)}번 들어있습니다.")

```

```python
lst = ['딸기', '바나나','수박','체리','포도']
ans = input("검색할 문자를 입력하세요 >>")

if ans in lst:
    print(f"{ans}는 리스트에 {lst.index(ans)}번째 인덱스에 들어있습니다.")
else:
    print(f"{ans}는 리스트에 들어있지 않습니다.")
```

```python
# if문 구조 (들여쓰기가 굉장히 중요- 안하면 오류뜸)
# ctrl + ] & ctrl + [ 로 들여쓰기 앞뒤로 자동맞출수있음.
num1 = 19
num2 = 15

if num1 > num2:
    print("num1이 더 큽니다")
    print('num2이 더 큽니다')
    if True :
        print("2번째 조건문")
    
print('if문이 종료되었습니다.')    
```

```python
num1 = int(input("첫 번째 정수 입력>>"))
num2 = int(input("두 번째 정수 입력>>"))

if num1>num2:
    print("첫 번째 정수가 더 큽니다.")
elif num1<num2:      # elif는 if문에 종속된거아니라 다시 들여쓰기빼줘야함
    print("두 번째 정수가 더 큽니다.")
else:
    print("두 수가 똑같습니다.")
```

```python
score = int(input("점수 입력 >>"))


if score <=100 and score>= 90:
    print(f"{score}점은 A학점 입니다.")
elif 90 < score >=80:
    print(f"{score}점은 B학점 입니다.")
elif score>=70:
    print(f"{score}점은 C학점 입니다.")
elif score>=60:
    print(f"{score}점은 D학점 입니다.")
else:
    print(f"{score}점은 F학점 입니다.")
```
